```python
import matplotlib.pyplot as plt

hours = ["Deep Sleep", "Light Sleep", "REM", "Awake"]
values = [3.2, 3.4, 0.6, 0.1]

plt.pie(values, labels=hours, autopct='%1.1f%%', colors=['#4B8BBE','#306998','#FFE873','#FFD43B'])
plt.title("My Sleep Cycle")
plt.show()
```

```python
import seaborn as sns
import numpy as np

data = np.random.rand(6,6)
sns.heatmap(data, cmap="YlGnBu")
plt.title("Workload Priority Heatmap")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "AI health climate space technology economy innovation AI ethics robotics sustainability"
wc = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import pandas as pd

hours = list(range(7,23))
stress = [40, 35, 50, 45, 55, 65, 60, 50, 45, 40, 35, 30, 25, 20, 15, 10]

df = pd.DataFrame({'Hour': hours, 'Stress': stress})
plt.plot(df['Hour'], df['Stress'], marker='o')
plt.title("My Stress Levels Throughout the Day")
plt.xlabel("Hour")
plt.ylabel("Stress Level")
plt.grid(True)
plt.show()
```
